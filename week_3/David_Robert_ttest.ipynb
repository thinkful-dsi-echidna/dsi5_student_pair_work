```
import pandas as pd
import numpy as np
from scipy import stats
#from statsmodels.stats.anova import AnovaRM
from statsmodels.stats.multicomp import pairwise_tukeyhsd
```

```
link = 'https://tf-assets-prod.s3.amazonaws.com/tf-curric/data-science/lung-volume.csv'
lung_data = pd.read_csv(link)
```

```
lung_data.head()
```

```
def unique(list1): 
    x = np.array(list1) 
    print(np.unique(x)) 
print(unique(lung_data['group']))
```

```
group_one = lung_data[lung_data['group'] == 1]
group_two = lung_data[lung_data['group'] == 2]
group_three = lung_data[lung_data['group'] == 3]
```

```
print(stats.ttest_rel(group_one['t1'], group_one['t2']))
print(stats.ttest_rel(group_two['t1'], group_two['t2']))
print(stats.ttest_rel(group_three['t1'], group_three['t2']))
```

```
pairwise_t2 = pairwise_tukeyhsd(lung_data['t2'] , lung_data['group'])
```
